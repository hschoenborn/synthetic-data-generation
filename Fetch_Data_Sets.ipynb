```python
import os
import requests
import pandas as pd

# Zugangsdaten
base_url = "https://research-external.infosim.net:4434/rest/measurements"
username = "pjs_wiinf_2024"
password = "UKhFWcqf4oeyb0HpgmHZ8cqWtrSe5G"

# Verzeichnis erstellen
output_base_dir = "measurements"
if not os.path.exists(output_base_dir):
    os.makedirs(output_base_dir)

def download_measurement(datasettype, measurement_id, unit, multiplier, file_format="json"):
    url = f"{base_url}/{datasettype}/{measurement_id}?last={unit},{multiplier}&type={file_format}"
    response = requests.get(url, auth=(username, password))
    if response.status_code == 200:
        # Verzeichnis f端r den Datensatztyp erstellen
        output_dir = os.path.join(output_base_dir, datasettype)
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        
        file_path = os.path.join(output_dir, f"{measurement_id}_last_{unit}_{multiplier}.{file_format}")
        with open(file_path, "wb") as file:
            file.write(response.content)
        print(f"Data for measurement ID {measurement_id} saved to {file_path}")
    else:
        print(f"Failed to download data for measurement ID {measurement_id}: {response.status_code} - {response.text}")

def download_all_measurements(measurement_ids, datasettypes, file_format="csv"):
    for datasettype in datasettypes:
        for measurement_id in measurement_ids:
            # Herunterladen der Daten f端r alle Zeit 4,6 = letzten 6 Monate
            download_measurement(datasettype, measurement_id, 4, 6, file_format)

# Beispielaufruf: Lade alle Messungen mit allen Typen f端r die letzten 5 Jahre (rawfulldata, aggregatedfulldata etc.)
measurement_ids = []  # Hier die IDs aus der Excel-Datei einf端gen
datasettypes = ["rawdata", "rawfulldata", "aggregateddata", "aggregatedfulldata"]

# Lade die IDs aus der Excel-Datei
excel_file_path = "all_interface_measurements_dmz_server.xlsx"
df = pd.read_excel(excel_file_path)
measurement_ids = df.iloc[:, 0].tolist()

# Lade alle Messungen
download_all_measurements(measurement_ids, datasettypes, file_format="csv")
```

```python
import glob

def find_and_copy_top_measurements(datasettypes, top_n=5):
    import shutil

    # Verzeichnis f端r h旦chste Traffic-Messungen erstellen
    highest_traffic_dir = "measurements_highest_traffic"
    if not os.path.exists(highest_traffic_dir):
        os.makedirs(highest_traffic_dir)

    for datasettype in datasettypes:
        # Erstelle Unterverzeichnis f端r jeden Datentyp
        type_dir = os.path.join(output_base_dir, datasettype)
        type_highest_traffic_dir = os.path.join(highest_traffic_dir, datasettype)
        if not os.path.exists(type_highest_traffic_dir):
            os.makedirs(type_highest_traffic_dir)

        # Finde alle CSV-Dateien im Verzeichnis des Datentyps
        csv_files = glob.glob(os.path.join(type_dir, "*.csv"))
        
        # Sortiere Dateien nach der Anzahl der Zeilen
        file_line_counts = [(file, sum(1 for line in open(file))) for file in csv_files]
        file_line_counts.sort(key=lambda x: x[1], reverse=True)
        
        # Kopiere die top_n Dateien
        for file, line_count in file_line_counts[:top_n]:
            shutil.copy(file, type_highest_traffic_dir)
            print(f"Copied {file} to {type_highest_traffic_dir}")

# Beispielaufruf: Finde und kopiere die 5 CSV-Messungen mit dem h旦chsten Traffic f端r jeden Typ
find_and_copy_top_measurements(datasettypes, top_n=5)
```
